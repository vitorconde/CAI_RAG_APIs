```python
import requests
import pandas as pd
import json
```

```python
url = "https://apps.spaguas.sp.gov.br/sibh/api/v2/measurements/now"

params = {
    "station_type_id": 1,
    "hours": 1,              # 1 até 72
    "show_all": "false",     # true ou false
    "serializer": "complete",# complete ou simplify
    "public": "true"         # true ou false
}

response = requests.get(url, params=params, timeout=30)
response.raise_for_status()

data = response.json()
```

```python
# transforma a lista de medições em DataFrame
df = pd.json_normalize(data["measurements"])

print(df.head())
print(df.columns.tolist())
```

```python
# salva o JSON completo
with open("dados_chuva.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("Arquivo salvo: dados_chuva.json")
```

```python
# ----------------------------
# 1) Quantidade de registros
# ----------------------------
num_registros = len(data.get("measurements", []))

print(f"Quantidade de registros: {num_registros}")

# ----------------------------
# 2) Tamanho do JSON (bytes)
# ----------------------------
json_bytes = response.content
size_bytes = len(json_bytes)
size_kb = size_bytes / 1024
size_mb = size_kb / 1024

print(f"Tamanho do JSON:")
print(f"- {size_bytes} bytes")
print(f"- {size_kb:.2f} KB")
print(f"- {size_mb:.4f} MB")
```

```python
import json
import pandas as pd

# 1) Ler JSON
with open("dados_chuva.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 2) Extrair lista de registros
records = data.get("measurements", data)

# 3) Converter para DataFrame
df = pd.json_normalize(records)

# 4) Salvar em CSV
df.to_csv("dados_chuva.csv", index=False, encoding="utf-8-sig")

print("CSV gerado: dados_chuva.csv")
```
